```python
import pandas as pd
import numpy as np
client_dictionary = {'name': ['Michael', 'Ana', 'Sean', 'Carl', 'Bob'], 
                     'second name': [None, 'Angel', 'Ben', 'Frank', 'Daniel'],
                     'birth place': ['New York', 'New York', 'Los Angeles', 'New York', 'New York'],
                     'age': [10, 35, 56, None, 28],
                     'number of children': [0, None, 2, 1, 1]}
df1 = pd.DataFrame(client_dictionary)
df2 = pd.DataFrame(client_dictionary)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.dropna(inplace=True) # When inplace is True, no output
```

```python
# Confirm it worked
df1.head()
```

```python
df2.dropna(inplace=False) # When inplace = False there is output.
```

```python
# Confirm it worked
df2.head()

# inplace = False returns a dataframe with the Nans dropped but...
# This is a new object.  It is not the original df2 dataframe.
```

```python
# inplace = True reassignes the new object to the original.  
# It is the same as writing - df2 = df2.dropna(inplace=False)
```
